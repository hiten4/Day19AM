Q1 — Difference between loops, list comprehension, and higher-order functions

Loops are used when we need step-by-step control over iteration. They are flexible and easy to debug but can be slower and more verbose.

List comprehension is a concise way to create lists. It is more readable and faster than loops for simple transformations.

Higher-order functions like map, filter, and reduce are used when applying a function to a collection. They make code more functional and compact.

Use:

Loops → complex logic or multiple conditions

List comprehension → simple transformations

Higher-order functions → functional style operations

Q2 — Flatten nested list and remove even numbers

```python
def flatten_and_filter(lst):
    return [x for sub in lst for x in sub if x % 2 != 0]


nested = [[1,2,3],[4,5,6],[7,8]]
result = flatten_and_filter(nested)
print(result)
```

Q3 — Hypothesis testing explanation

Hypothesis testing is a statistical method used to make decisions based on data.

Null hypothesis (H₀) assumes there is no effect or difference.
Alternative hypothesis (H₁) assumes there is a significant effect or difference.

The p-value measures how likely the observed data is if H₀ is true.
The significance level (α) is the threshold (commonly 0.05) to decide whether to reject H₀.

Example:
A company tests if a new feature increases sales.
H₀: No increase in sales
H₁: Sales increase

If p-value < 0.05, we reject H₀ and conclude the feature has an effect.
