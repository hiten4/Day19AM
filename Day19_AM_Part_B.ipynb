```python
import math
```

```python

def mean(data):
    return sum(data) / len(data)
```

```python

def variance(data):
    m = mean(data)
    return sum((x - m) ** 2 for x in data) / (len(data) - 1)
```

```python
def two_sample_t_test(sample1, sample2, alpha=0.05):
    n1 = len(sample1)
    n2 = len(sample2)

    m1 = mean(sample1)
    m2 = mean(sample2)

    v1 = variance(sample1)
    v2 = variance(sample2)

    se = math.sqrt((v1 / n1) + (v2 / n2))

    t_stat = (m1 - m2) / se

    critical_value = 1.96

    reject = abs(t_stat) > critical_value

    return {
        "t_statistic": t_stat,
        "reject_H0": reject,
        "alpha": alpha
    }


```

```python
# Example

A = [12, 23, 3, 34, 45, 56, 5]
B = [12, 1, 3, 1, 1, 2, 3, 4, 5, 3, 4]

result = two_sample_t_test(A, B)

print(result)
```
