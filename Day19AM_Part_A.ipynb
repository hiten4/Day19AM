# 1. Using loops

```python


def mean_loop(data):
    total = 0
    for x in data:
        total += x
    return total / len(data)


def median_loop(data):
    sorted_data = sorted(data)
    n = len(sorted_data)

    if n % 2 == 0:
        return (sorted_data[n//2 - 1] + sorted_data[n//2]) / 2
    else:
        return sorted_data[n//2]


def variance_loop(data):
    m = mean_loop(data)
    total = 0
    for x in data:
        total += (x - m) ** 2
    return total / len(data)
```

# 2. Using higher-order functions

```python
from functools import reduce

def mean_hof(data):
    return reduce(lambda a, b: a + b, data) / len(data)


def variance_hof(data):
    m = mean_hof(data)
    return reduce(lambda acc, x: acc + (x - m) ** 2, data, 0) / len(data)

```

# 3. Dataset operations

```python


marks = [45, 78, 88, 92, 67, 54, 73, 81]

avg = mean_loop(marks)

above_avg = list(filter(lambda x: x > avg, marks))

grades = [
    'A' if x >= 85 else
    'B' if x >= 70 else
    'C'
    for x in marks
]
```

# 4. Standard deviation


```python

import numpy as np

def std_dev(data):
    return (variance_loop(data)) ** 0.5


std_numpy = np.std(marks)


# 5. Two sample data

A = [12, 23, 3, 34, 45, 56, 5]
B = [12, 1, 3, 1, 1, 2, 3, 4, 5, 3, 4]

mean_A = mean_loop(A)
var_A = variance_loop(A)

mean_B = mean_loop(B)
var_B = variance_loop(B)

diff_means = mean_A - mean_B
```

# Output

```python


print(mean_A, var_A)
print(mean_B, var_B)
print(diff_means)
print(above_avg)
print(grades)
print(std_dev(marks), std_numpy)
```
